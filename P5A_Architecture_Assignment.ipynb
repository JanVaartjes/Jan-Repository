# P5A - Architecture
## Abstract
One of the innovative projects within Inbiolab B.V. is called “Smart Jacket”. The Smart Jacket is a small and soft elastic jacket that will be used in the neonatal intensive care unit. The Smart Jacket offers electrocardiogram and respiration electromyography monitoring. A missing but important vital sign in the Smart Jacket concept is blood oxygen saturation. Existing devices are not energy efficient enough to work on the Smart Jacket’s battery. Inbiolab B.V. is therefore asking the question: can a pulse oximeter be included in the Smart Jacket concept? To answer this question, a project that is called: “development of an energy efficient pulse oximeter” has been started. All aspects of developing a pulse oximeter are discussed In this project.

 

One of the most important differences with a conventional pulse oximeter is that the Smart Jacket will use a reflectance sensor instead of a transmittance sensor. This system is designed, implemented and tested together with an energy-efficient front-end integrated circuit. The testing phase was the most unpredictable. Based on an innovative technique, a new sensor was used on a different measuring site and with different settings than normal.

 

In response to the question “Is it possible to include an pulse oximeter in the Smart Jacket concept?”, a  “yes” can be answered based on the tests that were undertaken and the promising results that were achieved. This conclusion is supported by the estimated theoretical maximum power consumption of 2,75 mW and an accuracy of 1,2 %.

## Sensors used
#### Electrocardiography
To make an electrocardiogram (ECG) electrodes are placed on the skin, which will then record the electrical activity of the heart over a period of time. The depolarization of the heart during each heartbeat creates small electrical changes, which are measured by the electrodes.

#### Electromyography
Electromyography (EMG) works in the same way as electrocardiography, using electrodes placed on the skin. The difference is that an EMG focuses on the electrical potential generated by muscle cells of the skeletal muscles when these are activated. In the following chapters the EMG and ECG sensors are discussed in the same paragraphs, are the sensors used are equal, the only difference being the values of the human body they measure.

#### Oximetry
Using oximetry one can measure a person's oxygen saturation. This can be done using two methods: transmissive pulse oximetry or reflectance pulse oximetry. The project from which the abstract was taken focuses on reflectance pulse oximetry, which uses a red and infrared LED in combination with photodiodes, which absorbs the emission spectra of the red and infrared wavelengths in order to determine oxygen saturation.

## Boudjemaa & Forbes - Sensor Origin
#### Electrocardiography & Electromyography - Sensor
The electrodes used for electrocardiography and electromyography would fall in the sensor category, if the definition is taken in a broader sense in which several electrodes are used to measure changes in the electrical field on the human skin. When combining these measured changes one can visualize the heart rhythm and skeletal muscle activity.

#### Oximetry - Domain
The combinaton of leds and photodiodes fall in the domains category. LEDs are emitting red and infrared light, which have different wavelengths: the wavelength of red light is about 650nm, while infrared light (which is no longer in the visible spectrum) can have a wavelength between 1mm and 750nm. These LEDs emit light which is reflected by the human body, after which photodiodes sense the reflected light. Using these values a person's oxygen saturation can be determined. 

## Durrant-Whyte - Strategy
#### Electrocardiography & Electromyography - Complementary
In both electrocardiography and electromyography several different electrodes are used to measure the same process in the human body, heart rhythm and skeletal muscle activity respectively. In order to accurately measure the desired process and to get rid of any unwanted noise in the measured values, multiple sensors must be used and combined. When using multiple electrodes which complement eachother, the process can be measured with higher accuracy.

#### Oximetry - Complementary
The same goes for oximetry. Multiple LEDs and photodiodes are used in the measuring process in order to determine the actual values with a higher accuracy.

## Dasarathy - Input and Output Type
#### Electrocardiography & Electromyography - Data In, Features Out
Both electrocardiography and electromyography measure the change in electrical potential on the skin, which is registered as a simple yet very long list of continuously changing electrical values (data in). However, when combining all electrodes one can accurately determine what caused these changes and find structure in this raw data (features out).

#### Oximetry - Data In, Features Out
Again, oximetry has the same type of fusion as ECG and EMG. The photodiodes measure the reflected light originally emitted by the LEDs (data in), combining the sent and retrieved values in order to determine the oxygen saturation of one's blood (features out).
