```python
import httplib
import re
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

address='gratisweerdata.buienradar.nl'
c=httplib.HTTPConnection(address)
c.request('GET','/data/raintext?lat=53.01&lon=6.57')
original_sock=c.sock
content=c.getresponse().read()
#print('Output: \n{0}'.format(content))

#find desired values in string
rain = map(int,re.findall(('\d{3}'),content))
currenttime = re.findall(('\d{2}[:]\d{2}'),content)

#determine length of time string
Range=np.arange(len(currenttime))

#setup plot
x = np.linspace(0, 20)
Testdata = (1,5,7,8,9,4,2,4,8,3,5,4,6,8,1,3,2,4,9,5,6,5,1,7)
#line = plt.plot(Testdata[::3], linewidth=10, 'ro')
plt.plot(Range[::3],Testdata[::3], 'ro-')


# set the locations and labels of the xticks
plt.xticks(Range[::3],currenttime[::3])
plt.show()
```
