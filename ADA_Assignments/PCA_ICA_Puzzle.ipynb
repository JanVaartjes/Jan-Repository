# PCA Puzzle

```python
#Importing required libraries
#Importing image & setting image size

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from sklearn.decomposition import PCA

%matplotlib inline

image_height=12
image_width=120
image_shape=(image_height,image_width)
datafilename='PCA_challenge_for_Jan_Vaartjes.hdf5'

from tables import *

# Create Sample/Row types
class WordImage(IsDescription):
    idnumber  = Int64Col()      # Signed 64-bit integer
    image = Float64Col(image_shape)    # double (double-precision)

# Create PCA dataset for individual student
ImagesFile = open_file(datafilename, mode = "r")
ImagesTable = ImagesFile.root.Images.Images

pca = PCA()
```

```python
# Create empty array to store 1D images
DataArray = np.zeros((0,1440))
# print DataArray.shape

# Transforming images to 1D and stacking them
for x in ImagesTable.iterrows(0,3000):
    DataArray = np.vstack((DataArray,x['image'].ravel())) #gives 3000*1440 matrix for PCA               
```

```python
# Checking shape
# print DataArray.shape

#fit the model with DataArray
pca.fit(DataArray)

#obtain principal components
invPCA = pca.components_
AbsVal = np.abs(invPCA)

#reshaping image to original size
for i in range(10):
    plt.matshow(AbsVal[i:i+1].reshape((12,120)))

plt.show()
```
