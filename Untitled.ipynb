# P5A - Architecture
## Abstract
One of the innovative projects within Inbiolab B.V. is called “Smart Jacket”. The Smart Jacket is a small and soft elastic jacket that will be used in the neonatal intensive care unit. The Smart Jacket offers electrocardiogram and respiration electromyography monitoring. A missing but important vital sign in the Smart Jacket concept is blood oxygen saturation. Existing devices are not energy efficient enough to work on the Smart Jacket’s battery. Inbiolab B.V. is therefore asking the question: can a pulse oximeter be included in the Smart Jacket concept? To answer this question, a project that is called: “development of an energy efficient pulse oximeter” has been started. All aspects of developing a pulse oximeter are discussed In this project.

 

One of the most important differences with a conventional pulse oximeter is that the Smart Jacket will use a reflectance sensor instead of a transmittance sensor. This system is designed, implemented and tested together with an energy-efficient front-end integrated circuit. The testing phase was the most unpredictable. Based on an innovative technique, a new sensor was used on a different measuring site and with different settings than normal.

 

In response to the question “Is it possible to include an pulse oximeter in the Smart Jacket concept?”, a  “yes” can be answered based on the tests that were undertaken and the promising results that were achieved. This conclusion is supported by the estimated theoretical maximum power consumption of 2,75 mW and an accuracy of 1,2 %.


## Sensors used
#### Electrocardiography
To make an electrocardiogram (ECG) electrodes are placed on the skin, which will then record the electrical activity of the heart over a period of time. The depolarization of the heart during each heartbeat creates small electrical changes, which are measured by the electrodes.

#### Electromyography
Electromyography (EMG) works in the same way as electrocardiography, using electrodes placed on the skin. The difference is that an EMG focuses on the electrical potential generated by muscle cells of the skeletal muscles when these are activated.
