# libsoundannotator

Before we can start processing wav-files we need to install libsoundannotator from source.

We first need to clone the repository, I will assume you clone it here: 

~/Desktop/GIT_Workspace/libsoundannotator
    
It is further assumed you installed anaconde to abtain python.

We first navigate to the location of the repository:

```bash
cd ~/Desktop/GIT_Workspace/libsoundannotator
echo "Current directory:"
pwd
echo
echo 'Files in current directory:'
ls
```
    
The last command above should show among other things show the presence of:

```bash
libsoundannotator (directory)
setup.py (python setup script)
```

Fix some dependency issues by copying the following instruction to the commandline and executing them:

```bash
sudo apt-get install portaudio19-dev
conda install pip
pip install oauthlib
pip install pyaudio
```

Then we run the setup.py installation script:

```bash
python setup.py build install
```  
    
Create a ~/data before running tests:

```bash
mkdir ~/data
python setup.py test
```

The test will produce a lot of output but should contain a line like:

    ----------------------------------------------------------------------
    Ran 50 tests in 53.046s

If there are tests which failed these will be reported below this line:

    ----------------------------------------------------------------------
    Ran 50 tests in 41.980s

    FAILED (errors=14)

If all test pass you are ready to start to get the soundannotatordemo code working.

To process wav-files we can use the program soundAnnotator.py. This is a script written using the libsoundannotator library.

You can run it by providing the script name as an argument to python: python soundAnnotator.py. Before you enter this on the commandline you need to be aware that it needs additional arguments to be used properly.

The tract-feature extraction needs to use information about the filterbank used for creating the time-frequency representation, to do this **calibration** you first need to run soundAnnotator.py with the `-c` flag:

```bash
python soundAnnotator.py  -c --noofscales 100 --inputrate 44100 --outdir ~/data --decimation 2
```

The parameters for Resampling, and the parameters for the filterbank should be the same as used for actually processing data. Like the next example which processes example wav data: 

```bash
python soundAnnotator.py --PTNE  --ptnblockwidth 0.1 --noofscales 100 --inputrate 44100 --outdir ~/data --decimation 2 --wav /home/ronald/Desktop/GIT_Workspace/soundannotatordemo/soundannotatordemo/demodata/dares_g1
```

Note: the directory specified by `--outdir` should exist. This is also the location where you find the output. Rerunning the script will not overwrite teh output but it will create a new subdirectory. The name of this subdirectory is a combination of `GIT< SHA of commit><datetime>`. The presence of an identifier of the GIT commit should help you when you need to reproduce old results.
